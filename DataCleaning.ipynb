# LIBRARIES

```python
#DATA 
import pandas as pd
import numpy as np
#VISUAL
import matplotlib.pyplot as plt
import seaborn as sns
#IMPUTATION
from sklearn.impute import KNNImputer
```

# ENERGY DATA CLEANING

```python
EnergyData = pd.read_excel(f'OriginalData/ENERGYDATA.xlsx')
EnergyData
```

```python
EnergyData['DateTime'] = pd.to_datetime(EnergyData['Time Bucket (Asia/Calcutta)'], dayfirst=True, errors='coerce')
EnergyData.drop('Time Bucket (Asia/Calcutta)', axis=1, inplace=True)
EnergyData['Computer - kWatts'].replace("No CT", np.nan, inplace=True)
EnergyData['light + fan - kWatts '].replace("No CT", np.nan, inplace=True)
EnergyData
```

```python
EnergyData.dtypes
```

```python
print("total rows :",len(EnergyData))
print("nan computer :",len(EnergyData[EnergyData['Computer - kWatts'].isnull()]))
print("nan light fan :",len(EnergyData[EnergyData['light + fan - kWatts '].isnull()]))
```

```python
EnergyClean = EnergyData.dropna()
EnergyClean
```

# IAQ DATA CLEANING

```python
IAQ = pd.read_excel(f'OriginalData/Indoor Air Quality.xlsx')
IAQ
```

```python
IAQ['DateTime'] = IAQ['Location'].astype(str) + ' ' + IAQ['Unnamed: 1'].astype(str)
IAQ['DateTime'] = pd.to_datetime(IAQ['DateTime'], errors='coerce')
IAQ = IAQ.drop(['Location', 'Unnamed: 1'], axis=1)
IAQ['testo 160 IAQ_51616142 [ppm]'].replace("###", np.nan, inplace=True)
IAQ
```

```python
IAQ.dtypes
```

```python
print("total rows :",len(IAQ))
print("nan computer :",len(IAQ[IAQ['testo 160 IAQ_51616142 [ppm]'].isnull()]))
```

KNN IMPUTATION

```python
# Select only numeric columns for imputation
numeric_cols = IAQ.select_dtypes(include=[np.number]).columns
imputer = KNNImputer(n_neighbors=3)  # Choose k=5 (can be tuned)

# Apply KNN imputation
IAQ[numeric_cols] = imputer.fit_transform(IAQ[numeric_cols])
```

```python
IAQ
```

# OAQ DATA CLEANING

```python
OAQ = pd.read_excel(f'OriginalData/Outdoor Air Quality.xlsx')
OAQ
```

```python
OAQ['DateTime'] = OAQ['Location'].astype(str) + ' ' + OAQ['Unnamed: 1'].astype(str)
OAQ['DateTime'] = pd.to_datetime(OAQ['DateTime'], errors='coerce')
OAQ = OAQ.drop(['Location', 'Unnamed: 1'], axis=1)
OAQ['testo 160 IAQ_51616135_outdoor [ppm]'].replace("###", np.nan, inplace=True)
OAQ
```

```python
OAQ.dtypes
```

```python
print("nan computer :",len(OAQ[OAQ['testo 160 IAQ_51616135_outdoor [ppm]'].isnull()]))
```

KNN IMPUTATION

```python
# Select only numeric columns for imputation
numeric_cols = OAQ.select_dtypes(include=[np.number]).columns
imputer = KNNImputer(n_neighbors=3)  # Choose k=5 (can be tuned)

# Apply KNN imputation
OAQ[numeric_cols] = imputer.fit_transform(OAQ[numeric_cols])
```

```python
OAQ
```

# MERGE DATA

```python
AirQuality = pd.merge(OAQ, IAQ, on='DateTime', how='inner')
```

```python
Merged = pd.merge(AirQuality, EnergyClean, on='DateTime', how='inner')
```
